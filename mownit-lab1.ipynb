### Zadanie 1
Wykorzystaj funkcję finfo z biblioteki numpy aby określić precyzję i zakres różnych typów danych reprezentujących liczby zmienno-przecinkowe. Wyniki badania przedstaw w postaci zestawienia tabelrycznego.

### Zadanie 2 
Napisz funkcje, która dla dowolnej liczby zmiennoprzecinkowej w precyzji 64-bitowej (w standardzie IEEE 754) przedstawi w wersji dziesiętnej i binarnej składniki liczby: znak, wykładnik i mantysa. Można wykorzystać między innymi bibliotekę bitstring. Zbadaj jak reprezentowaną są poszczególne składniki tego formatu: zakres, sposób kodowania liczby etc. 

```python
import bitstring as bs
def float_info(floatn):
    fl = bs.BitArray(float=floatn, length=64)
    print ("|   Postac | Znak  |    Wykladnik     | Mantysa ")
    nestedList=[["Binarna",fl.bin[0],fl.bin[1:12],"1"+str(fl.bin[12:64])],
                ["Decimal",fl.bin[0],int(fl.bin[1:12],2),"1."+str(int(fl.bin[12:64],2))]]
    for item in nestedList:
        print ("|",item[0]," "*(6-len(item[0])),"|",
        item[1]," "*(4-len(str(item[1]))),"|",
        item[2]," "*(15-len(str(item[2]))),"|",
        item[3]," "*(24-len(str(item[3]))))

float_info(1.501)
```

### Zadanie 3 
Wykorzystaj funkcję nextafter() z biblioteki numpy (lub inny sposób) i sprawdź rozdzielczość typu float (jakie mają wartości najbliższe sąsiadujące liczby float). Zaprezentuj wynik korzystając z funkcji z Zadania 2. 

```python
number=1.501
eps=np.finfo(np.float64).eps
nestedList = [[eps,np.nextafter(number,number+1),
               eps+number]]
print("|   Rozdzielczość        |  Liczba najbliższa  | Liczba najbliższa korzystając z dodawania ")
for item in nestedList:
    print("|",item[0]," "*(12-len(str(item[0]))),"|",
              item[1]," "*(16-len(str(item[1]))),"|",
              item[2]," "*(23-len(str(item[2]))),"\n")
print (float_info(np.nextafter(number,number+1)))
```

Porównać reprezentację bitową liczby 1/3 dla typów float o różnej precyzji: float16, float32, float64. Sprawdź co się dzieje gdy zmienna o mniejszej precyzji konwertowana jest do wyższej.


```python
print("Liczba : 1/3")
print("|    Typ float  |        Liczba       |        Postac Binarna ")
nestedList=[["Float 16",(np.float16)(1.0/3.0),bs.BitArray(float=(np.float16)(1.0/3.0), length=32).bin[0:16]],
           ["Float 32",(np.float32)(1.0/3.0),bs.BitArray(float=(np.float32)(1.0/3.0), length=32).bin],
           ["Float 64",(np.float64)(1.0/3.0),bs.BitArray(float=(np.float64)(1.0/3.0), length=64).bin]]
for item in nestedList:
    print("|",item[0]," "*(12-len(str(item[0]))),"|",
              item[1]," "*(18-len(str(item[1]))),"|",
              item[2]," "*(23-len(str(item[2]))))
print("\n","Float 16 converted to float 32 :",np.float32(np.float16(1/3)));
```
