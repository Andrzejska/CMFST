# Interpolacja 

### Zadanie 1
Stablicuj następujące funkcje: sqrt(x), sin(x), x^3+2x w czterech punktach należących do przedziału 0 do 10.

```python
import math
import numpy as np
from scipy.interpolate import lagrange
import scipy.interpolate
import matplotlib.pyplot as plt 
sinT_x=[7*np.pi/2,np.pi/2,5*np.pi/2,3*np.pi/2]
sinT_y=[-1,1,1,-1]
sqrtT_x=[0,4,6,9]
sqrtT_y=[0,2,math.sqrt(6),3];
cubeT_x=[0,2,4,9]
cubeT_y=[0,12,72,747];
```

### Zadanie 2 
Napisz algorytm znajdujący wielomian interpolujący Lagrange dla powyższych stablicowanych funkcji.

```python
#Using Lagrange polynomial method for sin(x)
poly_sin=scipy.interpolate.lagrange(sinT_x,sinT_y);
print(poly_sin)
x=np.arange(0,10,0.1);
y=poly_sin(x);
plt.plot(x,y);
plt.show()
```

```python
#Using Lagrange polynomial for sqrt(x)
poly_sqrt=scipy.interpolate.lagrange(sqrtT_x,sqrtT_y);
print(poly_sqrt)
x=np.arange(0,10,0.1);
y=poly_sqrt(x);
plt.plot(x,y);
plt.show()
```

```python
import numpy as np
#Using Lagrange polynomial for cube(x)
poly_cube=scipy.interpolate.lagrange(cubeT_x,cubeT_y);
print(poly_cube)
x=np.arange(0,10,0.1);
y=poly_cube(x);
plt.plot(x,y);
plt.show()
```

### Zadanie 3
Porównaj wartość dokładną z wynikiem interpolacji dla punktów znajdujących się pomiędzy węzłami wielomianu (w połowie odległości) interpolującego. Oszacuj dokładność interpolacji. 


```python
print("Dla 4 węzlów interpolacji:" )
#Dla sin(x)
values=[np.pi/4,3*np.pi/4,5*np.pi/4]
actual_value=[];
poly_sinT=[];
for i in range (0,len(values)):
    actual_value.append(math.sin(values[i]));
    poly_sinT.append(poly_sin(values[i]));
error=np.absolute(np.subtract(actual_value,poly_sinT));
error=np.divide(error,actual_value);
error=np.average(error);
print("Niedokladnosc obliczenia sin(x) : ",error)

#Dla sqrt(x)
values=[2,5,7.5]
actual_value=[];
poly_sqrtT=[];
for i in range (0,len(values)):
    actual_value.append(math.sqrt(values[i]));
    poly_sqrtT.append(poly_sqrt(values[i]));
error=np.absolute(np.subtract(actual_value,poly_sqrtT));
error=np.divide(error,actual_value);
error=np.average(error);
print("Niedokladnosc obliczenia sqrt(x) : ",error)

#Dla x^3+2x
values=[1,3,6.5]
actual_value=[];
poly_cubeT=[];
for i in range (0,len(values)):
    actual_value.append((np.power(values[i],3))+values[i]*2);
    poly_cubeT.append(poly_cube(values[i]));
error=np.absolute(np.subtract(actual_value,poly_cubeT));
error=np.average(error);
print("Niedokladnosc obliczenia cube(x) : ",error)
```
